### 4.2 Dealing Missing Values

There are missing values in **Age**, **Sprint_40yd**, **Vertical_Jump**, **Bench_Press_Reps**, **Broad_Jump**, **Agility_3cone**, and **Shuttle**. In this notebook, we will impute these missing values using the mean of each column.

````python
# Columns to impute with the mean
cols_to_fill = ['Age', 'Sprint_40yd', 'Vertical_Jump', 'Bench_Press_Reps',
                'Broad_Jump', 'Agility_3cone', 'Shuttle']

# Impute both train and test using the means computed from the training data
for col in cols_to_fill:
    mean_value = train[col].mean()
    train[col] = train[col].fillna(mean_value)
    test[col] = test[col].fillna(mean_value)
````

Next, let's convert the categorical features into numerical values so they can be used as model imputs. Most machine learning models can only accept numerical data.

````python
# Label-encode the categorical features
label_encoders = {}
for c in ["Player_Type", "Position_Type", "Position"]:
    label_encoders[c] = LabelEncoder()
    train[c] = label_encoders[c].fit_transform(train[c].astype(str))
    test[c] = label_encoders[c].transform(test[c].astype(str))
````

## 5.Baseline Model

Now that the data has been prepared, we will use it as input to build a **baseline model**. A baseline model is a simple initial machine learning model that serves as a starting point.

In this notebook, we will use a **Random Forest** model, which is a tree-based method that makes predictions by combining many **decision trees**.

A decision tree is an intuitive model that repeatedly splits the data using simple “yes/no” questions. For example: “Is the player taller than 170 cm?” or “Is the player a kicker?” By following these conditions step by step, the model arrives at a final prediction.

However, using a single decision tree often leads to a model that fits the training data too closely and does not generalize well to new data—this is called **overfitting**. Random Forest addresses this by training dozens (or even hundreds) of decision trees, each on slightly different subsets of the data and features. It then combines their predictions—typically by averaging—resulting in a more stable and robust model.

To evaluate the model's performance, we will use **cross-validation**. This method splits the dataset into several groups (folds) and repeats training and evaluation multiple times.

In this notebook, we use **5-fold cross-validation**: the data is split into five folds, and each time we use one fold for validation and the remaining four folds for training. We repeat this process five times and take the average score as an estimate of model performance. Cross-validation is ueful because it provides a more reliable evaluation than a single train/validation split.

````python
# Split into features and target
X = train.drop(columns=["Drafted"])
y = train["Drafted"]

# Set up the model and cross-validation
model = RandomForestClassifier(
    n_estimators=100,
    max_depth=5,
    random_state=2025
)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Store scores
auc_scores = []

# Train and evaluate using Startified K-Fold cross-validation
for fold, (train_idx, valid_idx) in enumerate(skf.split(X, y)):
    print(f"Fold {fold + 1}")

    X_train, X_valid = X.iloc[train_idx], X.iloc[valid_idx]
    y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]

    # Train the model
    model.fit(X_train, y_train)

    # Predict and score
    y_valid_pred_proba = model.predict_proba(X_valid)[:, 1]
    auc = roc_auc_score(y_valid, y_valid_pred_proba)
    auc_scores.append(auc)
    print(f"  AUC: {round(auc, 4)}")

# Print mean AUC
mean_auc = np.mean(auc_scores)
print("\nAverage Validation AUC:", round(mean_auc, 4))
````

This model achieved an AUC score of **0.8115**.

### 7.1 Final model (the submission model)

The simple Random Forest above is a good *baseline*, but we can score noticeably higher on the leaderboard metric (**ROC-AUC**) with a stronger, fully-reproducible pipeline. Below we build the model that actually generates our submission.

**What this final model does (and why):**

1. **Feature engineering** — on top of `BMI`, we add domain features that describe athleticism and body composition:
   - `power_idx = Vertical_Jump + Broad_Jump`, `speed_power = Weight / Sprint_40yd`, `speed_score = Weight·200 / Sprint_40yd⁴` (the classic NFL "Speed Score"), `agility_diff`, `jump_ratio`, `bench_per_kg`.
   - **Missing-value indicators** (`*_isna`) and a `n_missing` count. Whether a player *skipped* a combine drill is itself informative (it correlates with draft outcome), so we keep that signal instead of erasing it.
   - **Position-relative z-scores**: each physical metric standardized *within the player's Position*, so "fast for a lineman" and "fast for a cornerback" are measured fairly.

2. **Leak-free target encoding** of the high-cardinality categoricals (`School`, `Position`, `Position_Type`, `Player_Type`). `School` has 236 levels — far too many to one-hot — so we replace each category with a **smoothed draft rate**. Crucially, the encoding statistics are computed **only on the training rows of each cross-validation fold**, so no target information leaks into validation/test.

3. **Model ensemble** — we average three complementary models by rank:
   - **CatBoost** using the categoricals *natively* (it handles `School`/`Position` without manual encoding),
   - **Random Forest** on the engineered + target-encoded features,
   - **Logistic Regression** (a linear model) for diversity.
   Because AUC only depends on the *ranking* of predictions, we combine the models with a **rank average**, which is robust and consistently beats any single model.

4. **Reproducibility** — every model uses a fixed `SEED`, as required by the competition rules.

With **5×5 repeated Stratified K-Fold** cross-validation this ensemble reaches an out-of-fold **AUC ≈ 0.839**, versus **≈ 0.813** for the baseline Random Forest.

````python
# --- Final model: setup, feature engineering, and helpers ---
# CatBoost is not pre-installed on Colab; install it quietly if it is missing.
try:
    from catboost import CatBoostClassifier, Pool
except Exception:
    import sys, subprocess
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "catboost"], check=True)
    from catboost import CatBoostClassifier, Pool

from pathlib import Path
import numpy as np
import pandas as pd
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from scipy.stats import rankdata

SEED = 42
PATH = Path.cwd() / "input"

# Reload the raw data so this section is independent of the earlier baseline
# preprocessing (which dropped/encoded columns in place).
train_full = pd.read_csv(PATH / "train.csv")
test_full = pd.read_csv(PATH / "test.csv")
y_full = train_full["Drafted"].astype(int).values

COMBINE = ['Sprint_40yd', 'Vertical_Jump', 'Bench_Press_Reps',
           'Broad_Jump', 'Agility_3cone', 'Shuttle']
CATF = ['School', 'Player_Type', 'Position_Type', 'Position']   # native cats for CatBoost
TE_COLS = ['School', 'Position', 'Position_Type', 'Player_Type']  # target-encoded cats


def base_fe(df):
    """Engineered numeric features (no target leakage)."""
    df = df.copy()
    for c in COMBINE + ['Age']:
        df[f'{c}_isna'] = df[c].isna().astype(int)
    df['n_missing'] = df[COMBINE].isna().sum(axis=1)
    df['BMI'] = df['Weight'] / (df['Height'] ** 2)
    df['power_idx'] = df['Vertical_Jump'] + df['Broad_Jump']
    df['speed_power'] = df['Weight'] / df['Sprint_40yd']
    df['agility_diff'] = df['Agility_3cone'] - df['Shuttle']
    df['jump_ratio'] = df['Vertical_Jump'] / df['Broad_Jump']
    df['bench_per_kg'] = df['Bench_Press_Reps'] / df['Weight']
    df['speed_score'] = (df['Weight'] * 200) / (df['Sprint_40yd'] ** 4)  # NFL Speed Score
    return df


TR = base_fe(train_full)
TE = base_fe(test_full)

ENG = ['BMI', 'power_idx', 'speed_power', 'agility_diff',
       'jump_ratio', 'bench_per_kg', 'speed_score', 'n_missing']
NUM = (['Year', 'Age', 'Height', 'Weight'] + COMBINE + ENG
       + [f'{c}_isna' for c in COMBINE + ['Age']])


def make_te_features(tr, ytr, other, alpha=20):
    """Smoothed, leak-free target encoding + Position-wise z-scores.
    All statistics are fit on `tr` (the fold's training rows) only."""
    tr, other = tr.copy(), other.copy()
    gmean = ytr.mean()
    feats = list(NUM)
    # Smoothed target encoding: blend each category's draft rate toward the
    # global mean (strong smoothing for rare schools/positions).
    for c in TE_COLS:
        cnt = tr.groupby(c).size()
        mean = tr.assign(_y=ytr).groupby(c)['_y'].mean()
        smooth = (mean * cnt + gmean * alpha) / (cnt + alpha)
        tr[f'{c}_te'] = tr[c].map(smooth).fillna(gmean)
        other[f'{c}_te'] = other[c].map(smooth).fillna(gmean)
        feats.append(f'{c}_te')
    # Position-relative z-scores (e.g. "fast for a lineman").
    for c in COMBINE + ['Weight', 'BMI']:
        st = tr.groupby('Position')[c].agg(['mean', 'std'])
        for d in (tr, other):
            m = d['Position'].map(st['mean'])
            s = d['Position'].map(st['std']).replace(0, np.nan)
            d[f'{c}_pz'] = (d[c] - m) / s
        feats.append(f'{c}_pz')
    return tr[feats], other[feats]


def rf_model():
    return Pipeline([
        ('imp', SimpleImputer(strategy='median')),
        ('rf', RandomForestClassifier(n_estimators=500, max_depth=7,
                                      min_samples_leaf=10, max_features='sqrt',
                                      random_state=SEED, n_jobs=-1))])


def lr_model():
    return Pipeline([
        ('imp', SimpleImputer(strategy='median')),
        ('sc', StandardScaler()),
        ('lr', LogisticRegression(C=0.5, max_iter=2000, random_state=SEED))])


def cat_model():
    return CatBoostClassifier(iterations=600, learning_rate=0.03, depth=5,
                              l2_leaf_reg=6, random_seed=SEED, verbose=0)


# Matrices for CatBoost's native categorical handling.
CAT_FEATS = NUM + CATF
Xcat_tr = TR[CAT_FEATS].copy()
Xcat_te = TE[CAT_FEATS].copy()
for c in CATF:
    Xcat_tr[c] = Xcat_tr[c].astype(str)
    Xcat_te[c] = Xcat_te[c].astype(str)
cat_idx = [Xcat_tr.columns.get_loc(c) for c in CATF]

print("Final-model features ready:", len(CAT_FEATS), "columns")
````

````python
# --- Train the ensemble with repeated Stratified K-Fold CV ---
# For every fold we (1) train each model on the fold's training rows, (2) record
# out-of-fold (OOF) validation predictions to estimate the leaderboard AUC
# honestly, and (3) predict the test set. Test predictions are averaged across
# all folds/repeats, then the three models are combined by rank averaging.

N_SPLITS, N_REPEATS = 5, 5
rkf = RepeatedStratifiedKFold(n_splits=N_SPLITS, n_repeats=N_REPEATS, random_state=SEED)

model_names = ['catboost', 'random_forest', 'logistic_regression']
oof = {m: np.zeros(len(TR)) for m in model_names}
oof_count = np.zeros(len(TR))
test_pred = {m: [] for m in model_names}

for fold, (trn_idx, val_idx) in enumerate(rkf.split(TR, y_full)):
    ytr = y_full[trn_idx]

    # 1) CatBoost with native categorical features
    cb = cat_model()
    cb.fit(Pool(Xcat_tr.iloc[trn_idx], ytr, cat_features=cat_idx))
    oof['catboost'][val_idx] += cb.predict_proba(Xcat_tr.iloc[val_idx])[:, 1]
    test_pred['catboost'].append(cb.predict_proba(Xcat_te)[:, 1])

    # 2) & 3) Random Forest and Logistic Regression on target-encoded features
    Xtr, Xval = make_te_features(TR.iloc[trn_idx], ytr, TR.iloc[val_idx])
    _, Xte = make_te_features(TR.iloc[trn_idx], ytr, TE)
    for name, mk in [('random_forest', rf_model), ('logistic_regression', lr_model)]:
        md = mk()
        md.fit(Xtr, ytr)
        oof[name][val_idx] += md.predict_proba(Xval)[:, 1]
        test_pred[name].append(md.predict_proba(Xte)[:, 1])

    oof_count[val_idx] += 1

# Average the repeated OOF predictions and report per-model + ensemble AUC.
rank01 = lambda p: rankdata(p) / len(p)
for m in model_names:
    oof[m] /= np.maximum(oof_count, 1)
    print(f"{m:20s} OOF AUC: {roc_auc_score(y_full, oof[m]):.4f}")

ensemble_oof = np.mean([rank01(oof[m]) for m in model_names], axis=0)
print(f"{'ENSEMBLE (rank avg)':20s} OOF AUC: {roc_auc_score(y_full, ensemble_oof):.4f}")

# Final test prediction: average each model across folds, then rank-average models.
test_mean = {m: np.mean(test_pred[m], axis=0) for m in model_names}
test_pred_proba_mean = np.mean([rank01(test_mean[m]) for m in model_names], axis=0)
````
